### information about columns

#### 1.sex-
0 represents female whereas 1 represents male
#### 2.trestbps(Resting blood pressure )-
High blood pressure = 140 mmHg or above
#### 3.chol(Serum cholestoral)-
Higher than 200 mg/dL is concerned
#### 4.fbs(Fasting blood sugar)-
0 represents < 120 mg/dL whereas 1 represents > 120 mg/dL
Higher than 126 mg/dL points to diabetes
#### 5.thalach-
Maximum heart rate achieved (bpm)
#### 6.exang(Exercise induced angina)
0:Negative
1: Positive
#### 7.oldpeak-
ST depression induced by exercise relative to rest
#### 8.ca-
Number of major vessels (0-3) colored by fluoroscopy
There is a clot if the blood vessel is not colored
#### 9.target( Heart Disease)
0 :Benign
1 :Malignant

## Data Analysis

### 1.male to female ratio 

```python
gender=data.sex.value_counts()
plt.pie(gender,labels=gender.index,autopct='%1.2f%%')
plt.title('Gender ratio')
plt.show()
```

```python
gender
```

### 2.benign to malignant ratio

```python
disease_ratio=data.target.value_counts()
plt.pie(disease_ratio,labels=disease_ratio.index,autopct='%1.2f%%')
plt.show()
```

### 3.distribution of heart disease according to gender

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='sex', data=data, hue='target')
plt.xlabel('Gender')
plt.ylabel('population')

plt.show()
```

```python
female=data[data.sex==0]
female1=female.target.value_counts()
plt.pie(female1,labels=female1.index,autopct='%1.2f%%')
plt.title('distribution of disease in females')
plt.show()
```

```python
male=data[data.sex==1]
male1=male.target.value_counts()
plt.pie(male1,labels=male1.index,autopct='%1.2f%%')
plt.title('distribution of disease in males')
plt.show()
```

### conclusion:


In case of females,the chances of having Malignant is higher.

### 4.distribution of disease according to age group

```python


plt.figure(figsize=(10, 5))
sns.countplot(x='age', data=data, hue='target')
plt.xlabel('Age group')
plt.ylabel('population')

plt.show()
```

### conclusion:
age group of 40-65 is more prone to heart diseases.
In case of age group of 40-55,malignant condition is more dominant,whereas in age group 55-65 benign is more dominant.

### 5.Distribution of chest pain types

```python
begign=data[(data.target==0)]
cp1=begign.cp.value_counts()
plt.pie(cp1,labels=cp1.index,autopct='%1.2f%%')

plt.title('benign')
plt.show()

Malignant=data[(data.target==1)]
cp2=Malignant.cp.value_counts()
plt.pie(cp2,labels=cp2.index,autopct='%1.2f%%')
plt.title('Malignant')
plt.show()
```

### conclusion:
In begign codition,individual mostly suffer from typical angina.

### correlations

```python
sns.relplot(x='age', y='cp', hue='target', data=data)
plt.xlabel('age')
plt.ylabel('chest pain types')
plt.show()
```

```python
sns.relplot(x='thalach', y='slope', hue='target', data=data)
plt.xlabel('Maximum heart rate achieved')
plt.ylabel(' slope')
plt.show()
```

individuals having Horizontal slope and Maximum heart rate achieved lower than 140 are more prone to Benign.

```python
sns.relplot(x='thalach', y='exang', hue='target', data=data)
plt.xlabel('Maximum heart rate achieved ')
plt.ylabel('Exercise induced angina')
plt.show()
```

individuals having negative exercise induced angina and high Maximum heart rate achieved are more prone to Malignant whereas individual having positive exercise induced angina and low Maximum heart rate achieved are prone to Benign.
